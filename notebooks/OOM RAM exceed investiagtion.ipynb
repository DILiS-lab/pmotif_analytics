```python
import os


random_graphs_path = "/home/timgarrels/masterthesis/datasets/random_graphs"
random_graphs = os.listdir(random_graphs_path)

command = "gtrieScanner -u -s 4 -f simple -g {random_graphs_path}/{dataset} -o motif_freq"
```

```python
random_graphs
```

```python
from typing import Dict


def run_gtrieScanner(dataset) -> Dict[str, int]:
    """Returns the motif freq lookup"""
    os.system(command.format(random_graphs_path=random_graphs_path, dataset=dataset))
    r = parse_gtrie_result("motif_freq", k=4)
    os.remove("motif_freq")
    return r
```

```python
"""Collect subgraph counts"""
from tqdm import tqdm
import pandas as pd
from pathlib import Path

out = Path("subgraph_size_4_counts.csv")


if not out.is_file():
    results = []
    for r in tqdm(random_graphs):
        results.append(run_gtrieScanner(r))


    df = pd.DataFrame(zip(random_graphs, [sum(r.values()) for r in results]), columns=["Graph", "Subgraph Count"]).set_index("Graph")
    df = df.sort_values(by="Subgraph Count")
    df.to_csv(out)
df = pd.read_csv(out)
df = df.set_index("Graph")
```

```python
import networkx as nx


graphs = [nx.read_edgelist(Path(random_graphs_path) / g) for g in df.index]

nodes = [len(g.nodes) for g in graphs]
edges = [len(g.edges) for g in graphs]
cc = [nx.average_clustering(g) for g in graphs]

df["nodes"] = nodes
df["edges"] = edges
df["cc"] = cc
df["density"] = [nx.density(g) for g in graphs]
df


df.sort_values(by="Subgraph Count")
```

```python
def graph_to_type(name: str):
    number, *rest = name.split("_")
    return "_".join(rest)



df["graph_type"] = list(map(graph_to_type, df.index))
```

```python
set(df["graph_type"])
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

plt.rcParams.update({'font.size': 18})

plt.rcParams['text.usetex'] = True

colors_for_graph_types = dict(zip(
    set(df["graph_type"]),
    list(mcolors.TABLEAU_COLORS.keys())[:len(set(df["graph_type"]))]
))


fig, ax = plt.subplots(1, 1, figsize=(8,7), dpi=1024)


# Plot Scatter
for graph_type in set(df["graph_type"]):
    data = df[df["graph_type"] == graph_type]
    
    for i, row in enumerate(data.to_dict(orient="records")):
            
        label = None
        if i == 0:
            label = graph_type
            
        marker = None
        if row["Subgraph Count"] > 1000000:
            marker = "x"
            
        
        ax.scatter(
            row["cc"],
            row["Subgraph Count"],
            color=colors_for_graph_types[graph_type],
            marker=marker,
        )
        

# Setup Legend
import matplotlib.lines as mlines

label_lookup = {
    'barabasi_albert_graph_m_1': r"\textsc{Barabási-Albert} $m_1$",
    'barabasi_albert_graph_m_2': r"\textsc{Barabási-Albert} $m_2$",
    'barabasi_albert_graph_m_3': r"\textsc{Barabási-Albert} $m_3$",
    'ferdos_renyi_graph_m_2000': r"\textsc{Erdős–Rényi}",
    'scale_free_graph_a_28_b_7_g_02': r"\textsc{scale-free\textsubscript{28}",
    'scale_free_graph_a_35_b_3_g_35': r"\textsc{scale-free\textsubscript{35}",
    'scale_free_graph_a_65_b_1_g_25': r"\textsc{scale-free\textsubscript{25}",
}

labels = []
handles = []

for graph_type, tex_label in label_lookup.items():
    
    labels.append(tex_label)
    handles.append(
        mlines.Line2D(
            [], [],
            color=colors_for_graph_types[graph_type],
            marker='o',
            linestyle='None',
            markersize=8,
            label=tex_label)
    )
    
labels.append("Aborted")
handles.append(
    mlines.Line2D(
        [], [],
        color="grey",
        marker='x',
        linestyle='None',
        markersize=8,
        label="Aborted")
)

labels.append("Finished")
handles.append(
    mlines.Line2D(
        [], [],
        color="grey",
        marker='o',
        linestyle='None',
        markersize=8,
        label="Finished")
)
    

ax.legend(handles=handles, labels=labels)


# Setup Axis
ax.set_xlabel("Average Clustering Coefficient")
ax.set_ylabel("Subgraph Count")

# ax.set_xscale("log")
ax.set_yscale("log")

fig.suptitle("Subgraph Count predicted by Avg. Clustering Coeficcient")

fig.tight_layout()
```
