```python
# List nodes which have a participation of above x stddevs
from statistics import mean, median, stdev


count_mean = mean(counts.values())
count_stdev = stdev(counts.values())
stdevs = 1
while 1:
    print(f"Mean + {stdevs} x stdevs")
    relevant = [n for n, count in counts.items() if count > count_mean + stdevs*count_stdev]
    print(relevant)
    print(len(relevant))
    if len(relevant) == 0:
        break
    stdevs += 1
```
