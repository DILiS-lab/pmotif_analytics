```python
# Load original data

import networkx as nx
import json
from pathlib import Path


original_graph = nx.read_edgelist(Path(os.environ["DATASET_DIRECTORY"]) / "karate_club_w_weights.edgelist")

with open(Path(os.environ["DATASET_DIRECTORY"]) / "karate_club_clusters.json", "r") as f:
    clusters = json.load(f)
    
with open(Path(os.environ["DATASET_DIRECTORY"]) / "karate_club_node_mapping.json", "r") as f:
    node_mapping = json.load(f)
```

```python
from pmotif_lib.result_transformer import ResultTransformer

r  = ResultTransformer.load_result(
    Path(os.environ["DATASET_DIRECTORY"]) / "karate_club.edgelist",
    Path(os.environ["EXPERIMENT_OUT"]),
    3,
)
```

```python
r.positional_metric_df
```

```python
r.positional_metric_df["mapped_nodes"] = r.positional_metric_df["nodes"].apply(lambda x: [node_mapping[n] for n in x])
```

```python
r.positional_metric_df
```

```python
node_to_cluster_lookup = {}
for cluster, nodes in clusters.items():
    for n in nodes:
        node_to_cluster_lookup[n] = cluster
```

```python
from tqdm import tqdm

for i, row in tqdm(
    r.positional_metric_df.iterrows(),
    total=r.positional_metric_df.shape[0],
    desc="Testing validity of metrics",
):
    graphlet_nodes = row["mapped_nodes"]
    
    subgraph = nx.induced_subgraph(original_graph, graphlet_nodes)
    internal_edge_weights = [
        e[2]["weight"]
        for e in subgraph.edges(data=True)
    ]
    if sorted(row["pInternalWeight"]) != sorted(internal_edge_weights):
        print("Internal Weights bad")
        break

    external_edges = set([
        tuple(sorted([u,v])) for u,v in original_graph.edges(graphlet_nodes)
    ]) - set([tuple(sorted([u,v])) for u,v in subgraph.edges])
    external_weights = [
        original_graph.get_edge_data(u, v)["weight"]
        for u,v in external_edges
    ]
    if sorted(row["pExternalWeight"]) != sorted(external_weights):
        print("external weights bad")
        break
        
    
    # pGivenModuleParticipation
    detected_modules = [list(clusters.keys())[index] for index in row["pGivenModuleParticipation"]]
    real_modules = [node_to_cluster_lookup[n] for n in graphlet_nodes]
    if sorted(set(detected_modules)) != sorted(set(real_modules)):
        print("modules bad")
        break
```
